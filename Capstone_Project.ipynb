```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset 
try:
    df = pd.read_csv('C:/Users/gokul/Downloads/Telegram Desktop/Entri Elevate/Python/Capstone Project/myexcel - myexcel.csv.csv')
except FileNotFoundError:
    print("Error: 'your_dataset.csv' not found. Please ensure the file exists and the path is correct.")
    exit()


# Preprocessing: Correcting the 'height' column
np.random.seed(42)  # for reproducibility
df['height'] = np.random.randint(150, 181, size=len(df))
```

```python
# Analysis Tasks

# 1. Distribution of employees across teams

team_counts = df['Team'].value_counts()
team_percentages = (team_counts / len(df)) * 100
print("Distribution of employees across teams:\n", team_counts)
print("\nPercentage split of employees across teams:\n", team_percentages)
```

```python
# 2. Segregate employees based on positions

positions = df['Position'].unique()
print("\nUnique employee positions:\n", positions)
```

```python
# 3. Predominant age group

age_groups = pd.cut(df['Age'], bins=range(18, 71, 5), right=False) # Adjust bins as needed
age_group_counts = age_groups.value_counts()
predominant_age_group = age_group_counts.idxmax()
print("\nPredominant age group:", predominant_age_group)
```

```python
# 4. Highest salary expenditure by team and position

team_salary_expenditure = df.groupby('Team')['Salary'].sum()
position_salary_expenditure = df.groupby('Position')['Salary'].sum()

highest_spending_team = team_salary_expenditure.idxmax()
highest_spending_position = position_salary_expenditure.idxmax()

print(f"\nTeam with highest salary expenditure: {highest_spending_team} ({team_salary_expenditure.max()})")
print(f"Position with highest salary expenditure: {highest_spending_position} ({position_salary_expenditure.max()})")
```

```python
# 5. Correlation between age and salary

correlation = df['Age'].corr(df['Salary'])
print(f"\nCorrelation between age and salary: {correlation}")
```

```python
# Visual Representation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()


plt.figure(figsize=(8, 6))
team_counts.plot(kind='bar')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()

```

```python
# Data Story
print("\nData Story:")
print("The analysis reveals insights about employee demographics, team structures and salary trends.")
print(f"Team distribution is as follows {team_counts}")
print("Salary expenditure varies by team and positions.")

print(f"Correlation between age and salary is {correlation}")
print("Further analysis can include exploring the impact of employee position, age group, and team, on their respective salaries.")
```
